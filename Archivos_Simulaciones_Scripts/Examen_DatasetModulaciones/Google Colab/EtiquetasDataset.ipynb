```python
import os
import csv

# Ruta al directorio principal del dataset
dataset_dir = '/content/drive/MyDrive/dataset'

# Nombre del archivo CSV de salida
csv_file = 'etiquetas.csv'

# Crear o abrir el archivo CSV
with open(csv_file, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['imagen', 'clase'])  # Cabecera del archivo CSV

    # Recorrer las carpetas del dataset
    for clase in os.listdir(dataset_dir):
        clase_path = os.path.join(dataset_dir, clase)

        if os.path.isdir(clase_path):
            # Recorrer las imágenes en la carpeta de la clase
            for imagen in os.listdir(clase_path):
                imagen_path = os.path.join(clase_path, imagen)

                # Escribir la fila en el archivo CSV
                writer.writerow([imagen_path, clase])
```

```python
import os
import csv

# Ruta al directorio principal del dataset
dataset_dir = '/content/drive/MyDrive/dataset'

# Recorrer las carpetas del dataset
for clase in os.listdir(dataset_dir):
    clase_path = os.path.join(dataset_dir, clase)

    if os.path.isdir(clase_path):
        # Nombre del archivo CSV de salida para la clase
        csv_file = f'{clase}.csv'

        # Crear o abrir el archivo CSV
        with open(csv_file, mode='w', newline='') as file:
            writer = csv.writer(file)
            writer.writerow(['imagen', 'clase'])  # Cabecera del archivo CSV

            # Recorrer las imágenes en la carpeta de la clase
            for imagen in os.listdir(clase_path):
                imagen_path = os.path.join(clase_path, imagen)

                # Escribir la fila en el archivo CSV
                writer.writerow([imagen_path, clase])
```
